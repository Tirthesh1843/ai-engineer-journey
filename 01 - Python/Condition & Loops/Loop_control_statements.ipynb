# Loop Control Statements

## Loop control statements are being used for terminating a loop in between when a condition is metor if you want to have some control over your iterating process or while iterating over series of numbers.

#### break ---> break is our first loop control statement which help you out to terminate the loop when certain condition is met.

```python
# print numbers from 0 to 50 but once you come across no. 18 terminate the loop. or do not print no. 18 or any number forward.
for i in range(0, 51):
    if i == 18:
        break
    else:
        print(i)
```

### Continue ---> continue is our secong control statement which helps in iterating over a series of numbers but helps us with skipping the values in between once we meet a condition.

```python
# User is asking for print nubers between 0 to 50 but  only print even numbers and skip odd numbers.
for i in range(0,51):
    if i%2 ==0:
        print(i)
    else:
        continue
```

```python
# print 0 to 20 but skip 5, 10, 15
for i in range(0,21):
    if i == 5 or i == 10 or i == 15:
        continue
    else:
        print(i)
```

### pass ---> Pass statement is used for doing nothing.

```python
# print the from 0 to 20 but for 17 do nothing
for i in range(0,21):
    if i == 17:
        pass
    else:
        print(i)
```

### f-String ---> f string will help us out with printing the numbers with the variable value present inside the "" so we can directly represent the variable value inside the "" message itself. Where variable value will be shown inside the {}.

```python
for i in range(6):
    for j in range(3):
        print(f"i value is: {i} and j value is: {j}")
```
