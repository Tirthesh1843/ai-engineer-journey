# Conditional Statement

## Conditional statements works like real life scenario where if certain condition is meeting than a statement of code or block of code gets executed.

```python
# Eligibility for Driving License on meeting certain conditions.

## Conditional staements---> If statement where if condition is meeting then a block of code gets executed.

age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible for driving license.")
else:
    print("You are not eligible for driving license.")
```

## Conditional Statements:

### if statement
### elif statement
### else statement

```python
# Figure out that who is the tallest among three friends.

friend_1_height = int(input("Enter the height of friend 1 in cm: "))
friend_2_height = int(input("Enter the height of friend 2 in cm: "))
friend_3_height = int(input("Enter the height of friend 3 in cm: "))

if friend_1_height > friend_2_height and friend_1_height > friend_3_height:
    print("Friend 1 is the tallest.")
elif friend_2_height > friend_3_height:
    print("Friend 2 is the tallest.")
else:
    print("Friend 3 is the tallest.")
```

## Nested Conditional Statement

### Nested conditional statements involve of statement inside another if statement.

```python
# compare height of three friends and print the taller one.
friend_1_height = int(input("Enter the height of friend 1 in cm: "))
friend_2_height = int(input("Enter the height of friend 2 in cm: "))
friend_3_height = int(input("Enter the height of friend 3 in cm: "))
if friend_1_height > friend_2_height:
    if friend_1_height > friend_3_height:
        print("Friend 1 is the tallest.")
    else:
        print("Friend 3 is the tallest.")
else:
    if friend_2_height > friend_3_height:
        print("Friend 2 is the tallest.")
    else:
        print("Friend 3 is the tallest.")
```

```python
# Check for leap year or not.
year = int(input("Enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```
