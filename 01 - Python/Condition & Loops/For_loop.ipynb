# Looping Statements:

## Looping statement can be used for iterating over a series of numbers or through iterables or Data structures or through a collection of data.

```python
# print the numbers from 1 to 5
print(1)
print(2)
print(3)
print(4)
print(5)
```

## range ()

### range () function can be used for printing or for working as our iterables.

### range() takes an input parameter and first parameter that the range will be taking as the input will be initial value that you want to print or you want to operate on and then next input after coma is an parameter as the final value and it can take third parameter as well which will be the step size and that is the difference between the iterating values.

```python
# print the numbers from 1 to 50
for i in range(1, 51, 1):
    print(i)
```

### In the range there is no neccessity for mentioning the step size and similarity if you want to print the numbers or you want to operate on numbers again there is no neccessity for passing the initial value as well if and only if your initial value is 0.

```python
# print the numbers from 0 to 100
for i in range(101):
    print(i)
```

```python
# print the numbers from 0 to to 100 with gape of 5
for i in range(0, 101, 5):
    print(i)
```

### By using range we can print numbers in descending order.


```python
# print numbers from 50 to 1 in reverse order
for i in range(50, 0, -1):
    print(i)
```

```python
# print the numbers from 50 to 0 with gape of 5 in reverse order
for i in range(50, -1, -5):
    print(i)
```

### In looping statements, We can iterate through iterables like string as well.

```python
string = "Tirthesh"
for i in string:
    print(i)
```

```python
# reverse the string using for loop
string = "Tirthesh"
for i in range(len(string)-1, -1, -1):
    print(string[i])
```

```python
# sum of n natural numbers 
n = int(input("Enter a number: "))
sum = 0
for i in range(n+1):
    sum += i
print(f"Sum of first {n} natural numbers is: {sum}")
```

```python
## print the prime numbers from 1 to 100
for i in range(2, 101):
    for j in range(2, int(i**(1/2))+1):
        if i % j == 0:
            break
    else:
        print(i)
```
