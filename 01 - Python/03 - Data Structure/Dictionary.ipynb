## Nested Dictionary

### Nested Dictionary will be used for keeping a dictionary inside another Dictionary.

```python
# I want to store multiple students data in a dictionary.
Nested_dictionary = {"student_1": {"name": "Tirthesh", "marks": 88, "Grade": "A"},
                        "Student_2": {"name": "Priya", "marks": 99, "Grade": "A"},
                        "Student_3": {"name": "Rohit", "marks": 77, "Grade": "B"},
                        "Student_4": {"name": "Vivek", "marks": 61, "Grade": "C"}}
```

```python
Nested_dictionary
```

```python
for outerKey,OuterValue in Nested_dictionary.items():
    print(f"OuterKey: {outerKey}, OuterValue: {OuterValue}")
    for key1,Value1 in OuterValue.items():
        print(f"Inner key is : {key1} and inner value is : {Value1}")
```

## Dictionary Comprehention

### It helps us with using looping statement along with Conditional Statement inside our Dictionary.

### Syntax for Dictionary Comprehention: 
### {expression for item in iterable if condition}

```python
comprehension = {x**2 for x in range(1,11)}
```

```python
comprehension
```

```python
Comprehensive_dictionary = {X:X**2 for X in range(1,11)}
```

```python
Comprehensive_dictionary
```

```python
even_square_dict = {X:X**2 for X in range(1,11) if X%2==0}
```

```python
even_square_dict
```
