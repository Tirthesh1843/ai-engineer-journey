# While Loop

## While loop is being used foir iterating over a serties of numbers when a condition is true and we have to initialize a value as wellwhile iterating.

```python
### By making the use uf while loop we can print the numbers from 0 to 50.
i = 0
while i <= 50:
    print(i)
    i += 1
```

```python
# print even numbers from 0 to 50 using while loop
i = 0
while i <= 50:
    if i % 2 == 0:
        print(i)
    else:
        pass
    i += 1
```

## Nested Loops

### It is going to be loop inside loop.

```python
print (0,0)
print (0,1)
print (0,2)
print (1,0)
print (1,1)
print (1,2)
print (2,0)
print (2,1)
print (2,2)
```

```python
# Nested loops can help us out for iterating over a series of numbers as with proper combination of numbers.

for i in range(5):
    for j in range(3):
            print(i, j)
```
