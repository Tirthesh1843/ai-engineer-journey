# Operators

## Operators helps in applying operations on given operands.

## Types of Operators:

1. Arithmatic Operator:

    1. Addition(+)
    2. Substraction(-)
    3. Multiplication(*)
    4. Division(/)
    5. Modulus (%)
    6. Floor (//)
    7. Exponent (a**b)

2. Comparison Operator:

    1. Greater than (>)
    2. Less than (<)
    3. Greater than or equal to (>=)
    4. Less than or equal to (<=)
    5. Equality check (==)
    6. Not Equal to (!=)

3. Logical Operators:

    1. and --> Whether if both conditions are true or not.
    2. or --> Whether if any of the operation is true or not.
    3. not --> negation or to inverse the result of given operation.
