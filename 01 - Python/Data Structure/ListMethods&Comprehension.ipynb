## List Methods:

#### 1. append() -> insert the value at the end of the list
#### 2. insert() -> insert the value at a specific index location in list.
#### 3. remove() -> remove the value from list that we are going to pass inside it.
#### 4. pop() -> remove the value from the list at specific index location.
#### 5. index() -> it provide the index of the value that we are going to pass insde it if that vale is present inside list.
#### 6. sort() -> it will sort the values in a assending manner.
#### 7. Reverse() -> it can reverse the values present inside our list that is value at last index will be at first index now.
#### 8. count() -> it counts the number of occurance that we are going to pass inside our list.
#### 9. clear() -> remove all elements in list and get empty list.

```python
Numbers = [1,12,234,2,5,6,7,6,32]
```

## enumerate() --> helps us with getting the index value along with the actual value that wee are storing inside the list.

```python
for index_value,value in enumerate(Numbers):
    print(index_value,value)
```

## len()

#### length of the list that means we will get the values as the total number of items we are storing inside the list.

```python
length = len(Numbers)
length
```

#### we are going tom use looping statements inside the list so that we can store the values inside the list in an iterable format and in list comprehension conditional statements are allowed as well. 

## Syntax for list Comprehension.

#### [expression for item in iterable if condition]

#### expression --> what we want to print.
#### for item in iterabel --> looping statement.
#### if Condition --> Conditional statement

```python
# store values from 1 to 50 inside the list.
list_items = [x for x in range(1,51)]
list_items
```

```python
list_square = [x**2 for x in range(1,7)] #7 is not included
list_square
```

```python
# square of even values between 1 to 18
even_square = [x**2 for x in range(1,19) if x%2==0]
even_square
```
