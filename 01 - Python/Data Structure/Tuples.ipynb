# Tuples

### Tuples are the second type of data Structure which can store a data in a ordered way just like Lists but Tuples are not mutable.

```python
a = tuple()
```

```python
type(a)
```

```python
tuple_values = (1,2,3,4,5)
print(tuple_values)
```

#### Tuples are like lists only which can store the data inside it in the ordered way but are going to be immutable

```python
tuple_values[1]
```

```python
tuple_values[1] = 22
```

#### Tuples are also capable of storing the data of any data type. 

```python
tuple_data = (1,2,3,4,"Tirthesh",4.76, True)
tuple_data
```

#### Tuples are going to be ordered or are going to be indexed.

```python
tuple_data[:]
```

```python
tuple_data[1:]
```

```python
tuple_data[::-1]
```

```python
tuple_data[2:5]
```
