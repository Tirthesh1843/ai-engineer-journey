```python
# We are creating a function to see how a translation is going to be there from English to hindi.

def translator(english_text):
    """This function is going to be used for converting English text to Hindi text."""
    english_text = english_text.lower()  # Convert input to lowercase for case-insensitive matching

    if english_text == "hello":
        print(f"The translation of '{english_text}' in Hindi is 'नमस्ते'.")
    elif english_text == "goodbye":
        print(f"The translation of '{english_text}' in Hindi is 'अलविदा'.")
    elif english_text == "help":
        print(f"The translation of '{english_text}' in Hindi is 'मदद'.")
    elif english_text == "thank you":
        print(f"The translation of '{english_text}' in Hindi is 'धन्यवाद'.")
    else:
        print(f"Sorry, the translation for '{english_text}' is not available in our dictionary.")
```

### For accessing our function we can simply provide the input to our function and we can simply call our function.

### Function calling is just like calling a waiter in restaurant where in just like we call the waiter and provide the input as the fod item and then waitor provide us with the food itself like in a similar way we can call function and give input and get requred output after calling the function only.

```python
# Calling a function ---> mantion name of the function and then provide the input.
English_input = input("Please enter an English word to translate to Hindi: ")
translator(English_input)
```

```python
# function to convert English to Hindi:

def translator_return(english_text):
    """This function is going to be used for converting English text to Hindi text."""
    english_text = english_text.lower()  # Convert input to lowercase for case-insensitive matching

    if english_text == "hello":
        Hindi_translation = "नमस्ते"
    elif english_text == "goodbye":
        Hindi_translation = "अलविदा"
    elif english_text == "help":
        Hindi_translation = "मदद"
    elif english_text == "thank you":
        Hindi_translation = "धन्यवाद"
    else:
        Hindi_translation = "Translation not available"
    
    return Hindi_translation
```
