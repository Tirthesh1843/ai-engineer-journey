# Function Examples

### Function are reusable by the we mean to say is we do not need to write same block of code again and again. This is sign of good programmer as well. As we don't need to write same code again and it can help us out with organizing code as well.

```python
# You are interviewer who is welcoming candidate each time with same greeting so you plan to use functions to avoid similar greeting again and again.

def greeting(user_name):
    """this function takes the input as the user_name and provides the out put as the greeting message."""
    print(f"Welcome to the interview, {user_name}! We are excited to have you here.")
    
candidate = input("Please enter your name: ")
greeting(candidate)
```

## Why to use Functions?
### It increase redability, Reusability and for organizing code as well

```python
## Creating a function that takes number from user and checks if the number is even or odd and returns the result.

def check_even_odd(number):
    """This function takes a number as input and checks if it is even or odd."""
    if number % 2 == 0:
        return "even number"
    else:
        return "odd number"

number = int(input("Please enter a number: "))
result = check_even_odd(number)
print(f"The number {number} is {result}.")
```

```python
print(check_even_odd(4))
```

```python
def even_odd_value(num):
    if type(num) != int:
        return "Please provide a valid integer input."
    elif num % 2 == 0:
        return "even number"
    else:
        return "odd number"
```

```python
even_odd_value("T")
```

```python
# isinstance()

def even_odd_value_instance(num):
    if isinstance(num, int):
        if num % 2 == 0:
            print(f"even number: {num}")
        else:
            print(f"odd number: {num}")
    elif isinstance(num, str):
        print(f"given input \"{num}\" is string:")
    else:
        print("Invalid input type. Please provide an integer.")
```

```python
even_odd_value_instance("asd")
```

```python
even_odd_value_instance(10.99)
```

```python
even_odd_value_instance(10)
```
