# List Methods

```python
# append ()

list_items = [1,2,3,4,5,6,7]
```

### append () ---> insert a value at the end of the list.

```python
list_items.append(8)
print(list_items)
```

```python
# insert ()
list_items.insert(3,9)
print(list_items)
```

```python
### remove ()
list_items.remove(4)
list_items
```
