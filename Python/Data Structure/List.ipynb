```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango", "orange", "pineapple"]
```

```python
print(fruits[2])
print(fruits[5])
```

```python
fruits[1:3]
```

## Lists are Mutable

#### By mentioning that list are mutable we mean to say that we can replace the values present at a specific location inside our list.

```python
animals = ["Dog", "Cat", "Tiger", "Lion", "Elephant", "Zebra"]
```

```python
animals[2]
```

```python
animals[2] ="Wolf"
```

```python
print(animals)
```

```python
animals[4] = "Monkey"
print(animals)
```

### By using looping statements we can iterate through list as well.

```python
for item in animals:
    print (item)
```
